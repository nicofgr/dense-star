```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

plt.rcParams['text.usetex'] = True

fig, ax1 = plt.subplots(2,2)
#-------------------------------------------------------
bagPressures = []
bagPressures.append(145)
bagPressures.append(155)
bagPressures.append(165)
bagPressures.append(175)
titles = []
titles.append('ud')
titles.append('uds')
titles.append('udsc')
titles.append('udscb')
titles.append('udscbt')
dirs = []
for bP in bagPressures:
    for i in titles:
        dirs.append('../cpp/output/' + i + 'B' + str(bP) +'.dat')
#print(dirs)
cols = []
cols.append('b')
cols.append('g-.')
cols.append('c--')
cols.append('r:')
cols.append('m')

for j in range(2):
    for k in range(2):
        x = []
        ax1[j,k].set_title("Presso e massa versus raio para B="+ str(bagPressures[k+2*j]) +"MeV")
        ax1[j,k].set_xlabel(r'$r$ (km)')
        ax1[j,k].set_ylabel(r'$p$ (MeV/f$m^3$)')
        ax1[j,k].set_xlim([0,12])
        ax1[j,k].xaxis.set_major_locator(MultipleLocator(2))
        ax1[j,k].xaxis.set_minor_locator(MultipleLocator(1))
        ax1[j,k].set_ylim([0,220])
        ax2 = ax1[j,k].twinx()
        ax2.set_ylabel(r'$m/M_\odot$')
        ax2.set_ylim([0,2.5])
        ax2.yaxis.set_major_locator(MultipleLocator(0.4))
        ax2.yaxis.set_minor_locator(MultipleLocator(0.2))
        
        for i in range(5):
            x.append([])
            with open(dirs[i+k*5+j*10], 'r') as f:   
                #print(dirs[i+k*5+j*10])
                lines = f.readlines()
                x[i].append([float(line.split()[0]) for line in lines])
                x[i].append([float(line.split()[1]) for line in lines])
                x[i].append([float(line.split()[2]) for line in lines])
        
            ax1[j,k].plot(x[i][0],x[i][1], cols[i], label=titles[i])
            ax2.plot(x[i][0],x[i][2], cols[i])
        
        ax1[j,k].legend(loc='upper center')
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()
```

```python
import math
c = 2.99792458e23
hbar = 6.58211915e-25
R0 = 1.47611
M0 = 8.9616e-4
B = pow(0.155,4)
n0 = 1.2277e-3
dr = 1e-3
mi = 1e-3
dn = 1e-2*n0
data = []


for nQuarks in range(2, 7):
    data.append([])
    data[nQuarks-2].append([])
    data[nQuarks-2].append([])
    data[nQuarks-2].append([])
    #Condies iniciais
    nc = 2*n0
    p0 = (nQuarks*pow(math.pi,2/3)/4)*pow(3*nc/nQuarks,4/3) - B
    
    while (nc/n0 < 15):
        r = 0
        m = 0
        mi = 1e-3
    
        p = p0
        n = nc
        e = 3*p + 4*B
        while ( p > 0 ):
            r = r + dr
            p = -R0*e*m*dr/pow(r,2)*(1+p/e)*(1+4*math.pi*pow(r,3)*p*M0/mi)/(1-2*R0*m/r)+p
            m = 4*math.pi*pow(r,2)*e*dr*M0/pow(hbar*c,3)+m
            mi = m
            e = 3*p + 4*B
        data[nQuarks-2][0].append(p0*1e3/pow(hbar*c,3))  # Presso
        data[nQuarks-2][1].append(m)  # Massa
        data[nQuarks-2][2].append(r)  # Raio
        nc = nc + dn
        p0 = (nQuarks*pow(math.pi,2/3)/4)*pow(3*nc/nQuarks,4/3) - B
print("Done")
```
